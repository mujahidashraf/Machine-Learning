```python
import numpy as np

# Sigmoid and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Dummy training data (X: 2 features, Y: 2 outputs)
# You can replace this with your actual data
mydata_X = np.array([[0.05, .1]])

mydata_Y = np.array([[0.01, .99]])

# Seed for reproducibility
np.random.seed(42)


```

```python
# Initialize weights and biases
input_neurons = 2
hidden_neurons = 2
output_neurons = 2

```

```python
# Weights
#W1 = np.random.rand(input_neurons, hidden_neurons)
#W2 = np.random.rand(hidden_neurons, output_neurons)

W1 = np.array([[0.15, .2],
               [.25,.30]])
W2 = np.array([[0.40, .45],
               [.50,.55]])

# Biases
b1 = np.array([[.35,.35]])
b2 = np.array([[.60,.60]])
```

```python
b2
```

```python

# Training the neural network
epochs = 20000
learning_rate = 0.6

for epoch in range(epochs):
    # Forward pass
    z1 = np.dot(mydata_X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Compute error
    error = mydata_Y - a2

    # Backpropagation
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
    d_W1 = np.dot(mydata_X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update weights and biases
    W2 += learning_rate * d_W2
    b2 += learning_rate * d_b2
    W1 += learning_rate * d_W1
    b1 += learning_rate * d_b1

    # print("\nFinal Weights and Biases after Training:")
    # print("W1 (Input to Hidden):")
    # print(W1)
    # print("\nb1 (Hidden Layer Biases):")
    # print(b1)
    # print("\nW2 (Hidden to Output):")
    # print(W2)
    # print("\nb2 (Output Layer Biases):")
    # print(b2)
    #break
    # Optionally print loss
    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch} Loss: {loss:.6f}")



# Test the model
print("\nFinal output after training:")
output = sigmoid(np.dot(sigmoid(np.dot(mydata_X, W1) + b1), W2) + b2)
print(output)
```

```python
import tensorflow as tf
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Load Iris dataset
iris = load_iris()
mydata_X = np.array([[0.05, .1]])

mydata_Y = np.array([[0.01, .99]])
X = mydata_X  # Features
y = mydata_Y

model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(shape=(X.shape[1],)),  # Input layer
    tf.keras.layers.Dense(2, activation='sigmoid'),  # Hidden layer with 10 neurons and ReLU activation
    tf.keras.layers.Dense(2, activation='sigmoid')  # Output layer with 3 neurons (for 3 classes) and softmax activation
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X, y, epochs=1, verbose=2)

test_loss, test_acc = model.evaluate(X, y)
print(f"Test Accuracy: {test_acc:.4f}")

# Make predictions
predictions = model.predict(X)
print(predictions)#
```
