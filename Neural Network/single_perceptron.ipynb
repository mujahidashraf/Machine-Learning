```python
import numpy as np


y = 1000
y = np.float64(y)

x1 = .01
x2 = .05

w1 = w2 = w3 = w4 = w5 = w6 = 1.0
b = 1.0
lr = 0.001

# Activation
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return 1.0 if x > 0 else 0.0

epochs = 20000
for e in range(epochs):
    # Forward pass
    z1 = w1*x1 + w2*x2 + b
    z2 = w3*x1 + w4*x2 + b
    h1 = relu(z1)
    h2 = relu(z2)
    yin = w5*h1 + w6*h2 + b
    yout = np.float64(yin)
    yout=relu(yout)

    # Error
    error = 0.5 * ((y - yout)**2)
    dy = yout - y

    dw5 = dy * relu_derivative(yout) *h1
    dw6 = dy * relu_derivative(yout)* h2
    db = dy * 1.0

    dz1 = dy * w5 * relu_derivative(h1)
    dz2 = dy * w6 * relu_derivative(h2)

    dw1 = dz1 * x1
    dw2 = dz1 * x2
    dw3 = dz2 * x1
    dw4 = dz2 * x2

    # Update weights
    w1 -= lr * dw1
    w2 -= lr * dw2
    w3 -= lr * dw3
    w4 -= lr * dw4
    w5 -= lr * dw5
    w6 -= lr * dw6
    b  -= lr * db  # update bias

    # Print every 1000 epochs
    if e % 100 == 0:
        print(f"Epoch {e}")
        print(f"yout  : {yout:.4f}")
        print(f"error : {error:.6f}")
        print(f"w5 w6     : {w5:.4f} {w6:.4f}")
        print(f"w1 w2     : {w1:.4f} {w2:.4f}")
        print(f"w3 w4     : {w3:.4f} {w4:.4f}")
        print("_" * 100)
        if error < 0.0001:
            break
```
