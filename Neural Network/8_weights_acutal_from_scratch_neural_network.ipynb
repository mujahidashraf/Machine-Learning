```python
import numpy as np

# Sigmoid activation and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(sig_x):
    return sig_x * (1 - sig_x)
```

```python
x1=0.05
x2=.1
y1=.01
y2=.99
w1=.15
w2=.2
w3=.25
w4=.3
w5=.4
w6=.45
w7=.5
w8=.55
b1=.35
b2=.6
# Learning rate
lr = 0.6
```

```python

max_epochs = 100000

# Training loop
for epoch in range(max_epochs):
  h1i = w1*x1+w2*x2 + b1
  h1 = sigmoid(h1i)
  #print(a1)

  h2i = w3*x1+w4*x2 + b2
  h2 = sigmoid(h2i)

  o1i=w5*h1+w6*h2+b2
  o1=sigmoid(o1i)
  o2i=w7*h1+w8*h2+b2
  o2=sigmoid(o2i)
  #print(o1)
  #print(o2)
  Eo1 = 0.5*((y1-o1)**2)
  Eo2= .5*((y2-o2)**2)
  #print(Eo1)
  #print(Eo2)
  Etotal =Eo1+Eo2
  #print(Etotal)
  dw5 = (o1-y1)*(o1*(1-o1))*h1
  #print(dw5)
  dw6 = (o1-y1)*(o1*(1-o1))*h2
  #print(dw6)

  dw7 = (o2-y2)*(o2*(1-o2))*h1
  #print(dw7)

  dw8 = (o2-y2)*(o2*(1-o2))*h2
  #print(dw8)

  deo1_h1=(o1-y1)*(o1)*(1-o1)*w5
  deo2_h1=(o2-y2)*(o2)*(1-o2)*w7
  dw1=(deo1_h1+deo2_h1)*h1*(1-h1)*x1
  #print(dw1)
  dw2 = (deo1_h1+deo2_h1)*h1*(1-h1)*x2
  #print(dw2)

  deo1_h2=(o1-y1)*(o1)*(1-o1)*w6
  deo2_h2=(o2-y2)*(o2)*(1-o2)*w8

  dw3= (deo1_h2+deo2_h2)*(h2)*(1-h2)*x1
  dw4= (deo1_h2+deo2_h2)*(h2)*(1-h2)*x2

  #print(dw3)
  #print(dw4)




  w1 -= lr * dw1
  w2 -= lr * dw2
  w3 -= lr * dw3
  w4 -= lr * dw4
  w5 -= lr * dw5
  w6 -= lr * dw6
  w7 -= lr * dw7
  w8 -= lr * dw8
  # print(w1)
  # print(w2)
  # print(w3)
  # print(w4)
  # print(w5)
  # print(w6)
  # print(w7)
  # print(w8)
  if epoch % 1000 == 0:

        print(f"Epoch {epoch} Loss: {Etotal:.8f}")
        print(f"y1^: {o1} y2^: {o2}")
        print(w1)
        print(w2)
        print(w3)
        print(w4)
        print(w5)
        print(w6)
        print(w7)
        print(w8)
        #print("weights updating")
        #print(W2)
        #print("input to hiiden")
        #print(W1)
        #print(loss)
        if Etotal<0.000009:
          break


```
